# Importing Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.show()
```

# Title

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

# Title
plt.title('Sample Graph')

plt.plot(x,y)
plt.show()
```

# X and Y Label

```python
x = [1,2,3,4,5]
# X Label
plt.xlabel('X AXIS', fontsize=20)

y = [6,7,8,9,10]
# Y Label
plt.ylabel('Y AXIS', fontsize=20)

plt.title('Sample Graph', fontsize=20)

plt.plot(x,y)
plt.show()
```

# Change Colour of Line

```python
x = [1,2,3,4,5]
plt.xlabel('X AXIS', fontsize=20)

y = [6,7,8,9,10]
plt.ylabel('Y AXIS', fontsize=20)

plt.title('Sample Graph', fontsize=20)

# Change the colour of the line
plt.plot(x,y, color='green', linestyle='dashed', marker='o')
plt.show()
```

# Change Ticks of the Axes

```python
x = [1,2,3,4,5]
plt.xlabel('X AXIS', fontsize=20)

# Change Ticks
plt.xticks([0,1,2,3,4,5,6])

y = [6,7,8,9,10]
plt.ylabel('Y AXIS', fontsize=20)

# Change Ticks
plt.yticks([6,7,8,9,10])

plt.title('Sample Graph', fontsize=20)

plt.plot(x,y,marker='o')
plt.show()
```

# Adding Legend

```python
x = [1,2,3,4,5]
plt.xlabel('X AXIS', fontsize=20)

y = [6,7,8,9,10]
plt.ylabel('Y AXIS', fontsize=20)

plt.title('Sample Graph', fontsize=20)

# Adding Legend
plt.plot(x,y, label='2X')
plt.legend()

plt.show()
```

# Short Hand for Line Format Change

```python
x = [1,2,3,4,5]
plt.xlabel('X AXIS', fontsize=20)

y = [6,7,8,9,10]
plt.ylabel('Y AXIS', fontsize=20)

plt.title('Sample Graph', fontsize=20)

#short Hand Technique
plt.plot(x,y, 'r--', label='2X')
plt.show()
```
